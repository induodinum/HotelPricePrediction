```python
import numpy as np
import pandas as pd
import pickle
from datetime import date

from sklearn.preprocessing import OneHotEncoder
```

```python
today = date.today()
today = today.isoformat()
```

```python
datapath = '../data/'

path_hotel_data = datapath + 'hotel_data.xlsx'
path_hotel_room_data = datapath + 'hotel_room_data.xlsx'
path_hotel_day5 = datapath + 'hotel_price_2019-04-05.pkl'
path_hotel_day6 = datapath + 'hotel_price_2019-04-06.pkl'
```

```python
day5_price_dict = pickle.load(open(path_hotel_day5,'rb'))
day6_price_dict = pickle.load(open(path_hotel_day6,'rb'))

hotel_price_day5 = pd.DataFrame(day5_price_dict)
hotel_price_day6 = pd.DataFrame(day6_price_dict)

hotel_data = pd.read_excel(path_hotel_data)
hotel_room_data = pd.read_excel(path_hotel_room_data)
hotel_price = pd.concat([hotel_price_day5,hotel_price_day6],ignore_index=True)
```

```python
hotel_price.shape
```

```python
del hotel_price['Hotel_prices_standard']
#hotel_price.columns = ['hotel_name','hotel_room_Type_Names','Hotel_prices','hotel_room_Benefits','Time_collect']
merged_data = pd.merge(hotel_data,hotel_room_data,how='inner',on='hotel_id_in_our_database')
count_null = merged_data.isna().sum()
count_null[count_null != 0]
```

```python
merged_data = merged_data.dropna(subset=['Room Capacity','hotel_star'])
merged_data.loc[:,['view']] = merged_data.loc[:,['view']].fillna('unknown')
count_null = merged_data.isna().sum()
count_null[count_null != 0]
```

```python
merged_data.rename(columns={'hotel_room_Type_Names': 'room_type', 
                            'hotel_room_Benefits': 'room_benefits'
                           }, 
                   inplace=True)
```

```python
hotel_price.rename(columns={'Hotel_name': 'hotel_name', 
                            'Hotel_roomtype': 'room_type', 
                            'Hotel_prices': 'room_price', 
                            'Hotel_benefits': 'room_benefits', 
                            'Time_collect': 'time_collected'
                           }, 
                   inplace=True)
```

```python
merged_data_n = pd.merge(merged_data,hotel_price, how='inner', 
                         on=['hotel_name', 'room_type', 'room_benefits']) # join the data
```

```python
cols = ['hotel_name', 'hotel_location',
        'hotel_star', 'hotel_user_Ratings',
        'hotel_room_Type_Names', 'hotel_room_Benefits',
        'Hotel_prices', 'Time_collect']
merged_data_c = merged_data_n[cols]
```

```python
merged_data_c
```

```python
cleaned_data = merged_data_c.copy()
cleaned_data['hotel_location'] = cleaned_data['hotel_location'].astype(str)

cleaned_data['district'] = cleaned_data['hotel_location'].str.extract('(.*), Bangkok')
cleaned_data['district'] = cleaned_data['district'].apply(lambda x: x.split(',')[-1])
cleaned_data['district'] = cleaned_data['district'].apply(lambda x: x.strip(' '))
```
