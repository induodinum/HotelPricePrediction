```python
driver = webdriver.Chrome('/Users/admin/Downloads/chromedriver')
driver.get(base_url)

assert "Booking.com" in driver.title

time.sleep(5)

elem = driver.find_element_by_name("ss")

elem.clear()
elem.send_keys("Bangkok")

calendar_xpath = "//div[@class='xp__dates-inner xp__dates__checkin']"
date_1_xpath = "//td[@class='bui-calendar__date' and @data-date='2019-02-13']"
date_2_xpath = "//td[@class='bui-calendar__date' and @data-date='2019-02-14']"

chkin_elem = driver.find_element_by_xpath(calendar_xpath)

actions = ActionChains(driver)
actions.move_to_element(chkin_elem)
actions.click(chkin_elem)
actions.perform()

time.sleep(2)

chkin_date_elem = driver.find_element_by_xpath(date_1_xpath)
actions.move_to_element(chkin_date_elem)
actions.click(chkin_date_elem)
actions.perform()

time.sleep(1)

chkout_date_elem = driver.find_element_by_xpath(date_2_xpath)
actions.move_to_element(chkout_date_elem)
actions.click(chkout_date_elem)
actions.perform()

elem.send_keys(Keys.RETURN)

time.sleep(3)
driver.implicitly_wait(2)

hname_path = "//a[@class='hotel_name_link url']"
hname_list = driver.find_elements_by_xpath(hname_path)
hlink = [ele.get_attribute('href') for ele in hname_list]
```
