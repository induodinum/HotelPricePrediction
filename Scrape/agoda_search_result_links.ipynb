```python
def chk_inout():
    
    chk_list = []
    n = 5
    STR_LEN = len('2019-02-09')
    
    t_date = datetime.now()
    
    for i in range(1,n):
        
        chkin = (t_date + timedelta(days=i))
        chkout = (t_date + timedelta(days=i+1))
        
        chkin = chkin.strftime('%a %b %d %Y')
        chkout = chkout.strftime('%a %b %d %Y')
        
        chk_list.append(chk(chkin, chkout))

    return chk_list
```
