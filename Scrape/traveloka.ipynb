```python
# x = soup.find(class_='_3HD0_ tvat-roomName' )
# x
def click_etc() :
    room_xpath = '//div[@class="_1SviU tvat-toggleAllRooms"]'
    room_elements = driver.find_elements_by_xpath(room_xpath) 
    for i in range(len(room_elements)) :
        room_elements[i].click()
        time.sleep(3)
    time.sleep(3)
    each_hotel_raw_html = driver.execute_script("return document.documentElement.outerHTML")
    soup = BeautifulSoup(each_hotel_raw_html,'lxml')
    return soup
# soup.findAll(class_='_3HD0_ tvat-roomName' )
```

```python
def check_none(name,List) :
    if name is not None :
        List.append(name.get_text())
    else :
        List.append('None')
    return List
```

```python
def data_each_hotel() :
    click_etc()
    driver.implicitly_wait(3)
    time.sleep(2)
    each_hotel_raw_html = driver.execute_script("return document.documentElement.outerHTML")
    soup = BeautifulSoup(each_hotel_raw_html,'lxml')
    time.sleep(5)
    roomName = [] ; bedType = [] ; refund = [] ; breakfast = [] ; Wifi = [] ; realPrice = [] ; disPrice = [] ; 
    save =[] ; Tax = [] ; hotelName = [];
    
    for i in soup.findAll(class_='yob74' ) :
        hotel_name = soup.find(class_="_3nicH tvat-hotelName")
        hotelName.append(hotel_name.get_text())
        room_name = i.find(class_='_3HD0_ tvat-roomName')
        roomName.append(room_name.get_text())
        bed_type = i.find(class_="_24P5a tvat-roomBedType")
        check_none(bed_type,bedType)
        refun = i.find(class_="_216CV")
        check_none(refun,refund)
        rice = refun.find_next(class_="_216CV")
        check_none(rice,breakfast)
        wifi = rice.find_next(class_="_216CV")
        check_none(wifi,Wifi)
        rp = i.find(class_="Y5Dqs")
        check_none(rp,realPrice) 
        dp = i.find(class_='_2qcja tvat-roomPrice')
        check_none(dp,disPrice)
        s = i.find(class_="_2ZXo8")
        check_none(s,save)
        tax = i.find(class_="_1pqB-")
        check_none(tax,Tax)
        df = pd.DataFrame(np.array([hotelName,roomName,bedType,refund,breakfast,Wifi,realPrice,disPrice,save,Tax]),\
                 index=['hotelName','roomName','bedType','refund','breakfast','Wifi','realPrice','disPrice','Save','Tax']).T
    return df
```
