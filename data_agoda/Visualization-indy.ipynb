```python
bed_type_info = {'futon':0.5,'futons':0.5,'bunk bed':1,'sofa bed':2,'semi single bed':3,'single bed':4,'semi double bed':5,'double bed':6,'queen bed':7,'king bed':8,'super king bed':9}
import re
for s in ['bunk beds, ','double beds, ','double bed, ','king bed, ','king beds','queen bed, ','semi double bed, ','semi double beds, ','single bed, ','single beds, ','super king bed, ']:
    f = visualize_data[visualize_data['bed_type'] == s]['Features']
    for i,index in zip(f,f.index):
        k = i.split('&&')[0]
        num_bed = 0
        max_score = 0
        try:
            for j in k.split(' and '):
                num_bed += int(re.search('(.) .*',j).group(1))
                num_bed_each = int(re.search('(.) .*',j).group(1))
                if num_bed_each == 1:
                    bed_type_list = re.search('[1-9] (.*)',j).group(1)
                else:
                    bed_type_list = re.search('[1-9] (.*)s',j).group(1)
                if bed_type_info[bed_type_list] > max_score:
                    max_score = bed_type_info[bed_type_list]
                    best_bed_type_list = bed_type_list
            visualize_data.bed_type.loc[index] = best_bed_type_list
            visualize_data.num_of_beds.loc[index] = num_bed
        except KeyError:
            pass
```

```python
#### clean an error ####
Data = visualize_data.Features[visualize_data.bed_type == 'contains ,']
Index = visualize_data.Features[visualize_data.bed_type == 'contains ,'].index
for Features,Indice in zip(Data,Index):
    max_score_overall = 0
    beds = Features.split('&&')[0]
    i = beds.split(' or ')
    for j in i:
        if 'and' not in j:
            num_bed = int(re.search('([1-9]) .*',j).group(1))
            if num_bed == 1:
                bed_type = re.search('[1-9] (.*)',j).group(1)
            else:
                bed_type = re.search('[1-9] (.*)s',j).group(1)
        else:
            num_bed = 0
            max_score = 0
            for k in j.split(' and '):
                num_bed += int(re.search('(.) .*',k).group(1))
                num_bed_each = int(re.search('(.) .*',k).group(1))
                if num_bed_each == 1:
                    bed_type_each = re.search('[1-9] (.*)',k).group(1)
                else:
                    bed_type_each = re.search('[1-9] (.*)s',k).group(1)
                if bed_type_info[bed_type_each] > max_score:
                    max_score = bed_type_info[bed_type_each]
                    bed_type = bed_type_each
        if bed_type_info[bed_type] > max_score_overall:
            max_score_overall = bed_type_info[bed_type]
            best_bec_type = bed_type
            num_bec = num_bed
    visualize_data.bed_type.loc[Indice] = best_bec_type
    visualize_data.num_of_beds.loc[Indice] = num_bec
```
